```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

                                ADS ASSIGNMENT 2
           1. A breakdown of distance from home by job role and attrition.

```python
data = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data2 = data[['Attrition', 'JobRole', 'DistanceFromHome', 'MonthlyIncome', 'Education' ]]
data2
```

```python
data2.groupby(['DistanceFromHome', 'JobRole']).count().reset_index()
```

```python
Attrited_yes = data2[['DistanceFromHome', 'JobRole','Attrition' ]][data['Attrition'] == 'Yes']
Attrited_yes.groupby(['DistanceFromHome','JobRole']).count().head(10)
```

```python
Attrited_no = data2[['DistanceFromHome', 'JobRole','Attrition' ]][data['Attrition'] == 'No']
Attrited_no.groupby(['DistanceFromHome', 'JobRole',]).count().head(10)
```

                        2. Compare average monthly income by education and attrition

```python
data['Education'].value_counts()
```

```python
data_monthlyincome = data2[['Attrition', 'MonthlyIncome', 'Education']]
data3 = data_monthlyincome.groupby(['Attrition','Education']).mean().reset_index()
data3
```

                                    ADS ASSIGNMENT 3

```python
plt.figure(figsize=(16,5))
sns.barplot(x="JobRole", y="DistanceFromHome" ,hue="Attrition", data=data2,)
plt.xlabel("Job Role")
plt.ylabel("Distance From Home")
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Education", y="MonthlyIncome" ,hue="Attrition", data=data3)
plt.xlabel("Education")
plt.ylabel("Average MonthlyIncome")
```
